```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import decode_predictions
from IPython.display import Image
```

```python
# Change directory to the folder containing images
os.chdir("/content/drive/MyDrive/BTK/day9/resimler")
path="/content/drive/MyDrive/BTK/day9/resimler"
file_names=os.listdir()
file_names
```

```python
# Display an image
Image(os.path.join(path, file_names[2]))
```

```python
# Function to preprocess the image
def img_preprocessing(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = tf.keras.applications.resnet50.preprocess_input(img)
    return img
```

```python
# Load the ResNet50 model pre-trained on ImageNet
model=ResNet50(weights="imagenet")
```

```python
# Loop through the files and predict
for file in file_names:
    img_path = os.path.join(path, file)
    img2 = img_preprocessing(img_path)
    pred = model.predict(img2)
    print(img_path, decode_predictions(pred, top=1)[0])
```
